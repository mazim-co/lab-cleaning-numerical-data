```python
#1. import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#2. load data
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df.head()
```

```python
#3. details about shape and info
customer_df.info()
```

```python
#4. rename cols with lower case and replace space with underscore
cols = []
for i in range(len(customer_df.columns)):
    cols.append(customer_df.columns[i].lower().replace(' ', '_'))

customer_df.columns = cols
```

```python
#5. change effective to date column to datetime format
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
```

```python
#6. checking for null values
customer_df.isnull().sum()/len(customer_df)
```

```python
#7. function that differentiates between continuous and discrete variables

customer_df.info()
```

```python
customer_df.describe(include='object')
```

```python
customer_df.describe(include='number')
```

```python
numcols = customer_df.select_dtypes(np.number)
numcols_list = list(numcols.columns)
discrete = []
continuous = []
for col in numcols_list:
    item_counts = customer_df[col].value_counts()
    print(col, ': ', len(item_counts))
    if len(item_counts) > 100:
        continuous.append(col)
    else:
        discrete.append(col)
discrete.append('effective_to_date')
```

```python
print(discrete)
print(continuous)
```
